```python
# preprocess_celeba.py
import os
import torch
from torchvision import transforms
from PIL import Image
from tqdm import tqdm

IMG_DIR = "img_align_celeba"
OUT_FILE = "celeba_tensor.pt"
LIMIT = 10000  # Change if you want full dataset

transform = transforms.Compose([
    transforms.Resize(64),
    transforms.CenterCrop(64),
    transforms.ToTensor(),
])

image_files = [f for f in os.listdir(IMG_DIR) if f.endswith(('.jpg', '.png'))]
image_files = image_files[:LIMIT]

images = []

print(f"Preprocessing {len(image_files)} images...")

for f in tqdm(image_files):
    path = os.path.join(IMG_DIR, f)
    img = Image.open(path).convert('RGB')
    img = transform(img)
    images.append(img)

tensor = torch.stack(images)
torch.save(tensor, OUT_FILE)

print("Saved:", OUT_FILE)
```
